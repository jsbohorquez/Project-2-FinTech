```python
import numpy as np 
from sklearn.svm import SVR
import matplotlib.pyplot as plt
import pandas as pd
import hvplot.pandas 
import holoviews as hv 
import yfinance as yf
from pathlib import Path
```

```python
sp500 = yf.Ticker('^GSPC')

```

```python
sp500.info
```

```python
# get historical market data
sp500_df = sp500.history(period='max')
sp500_df = sp500_df.drop(columns=['Dividends', 'Stock Splits'], axis=1)
sp500_df.tail()
```

```python
sp500_df.to_csv('S&P_500.csv',float_format='%.2f')
```

```python
SP500 = pd.read_csv(
    Path("S&P_500.csv"), 
    index_col='Date', 
    infer_datetime_format=True, 
    parse_dates=True
)

# Review the DataFrame
SP500.tail(20)
```

```python
# Filter the date index and close columns
SP500 = SP500.loc[:, ["Close"]]

# Use the pct_change function to generate  returns from close prices
SP500["Actual Returns"] = SP500["Close"].pct_change()

# Drop all NaN values from the DataFrame
SP500 = SP500.dropna()

# Review the DataFrame
display(SP500.tail(20))
```

```python
# windows
short_window = 50
long_window = 100

# fast and slow SMA (50 and 100 days)
SP500['SMA_Fast'] = SP500['Close'].rolling(window=short_window).mean()
SP500['SMA_Slow'] = SP500['Close'].rolling(window=long_window).mean()

SP500 = SP500.dropna()

# Review the DataFrame
display(SP500.tail(20))
```

```python
sp500_shortma = SP500.drop(columns = ['Actual Returns', 'SMA_Fast', 'Close'])
sp500_shortma = sp500_shortma.reset_index()
display(sp500_shortma)
```

```python
#import train_test_split function
from sklearn.model_selection import train_test_split

#ASK IF CORRECT FOR X, y
#set targets and features
featuresx = sp500_shortma[['Date']]

targety = sp500_shortma[['SMA_Slow']]

```

```python
from sklearn.preprocessing import StandardScaler
StdS_X = StandardScaler()
StdS_y = StandardScaler()
featuresx = StdS_X.fit_transform(featuresx)
targety = StdS_y.fit_transform(targety)
```

```python
training_features, testing_features, training_targets, testing_targets = train_test_split(featuresx, targety)

```

```python
#initialting the model
svr_model = SVR(kernel = 'linear', C = 10)
```

```python
#fitting the data to the model
svr_model.fit(training_features, training_targets)
```

```python
pred_y = svr_model.predict(training_features)
print(pred_y)
```
